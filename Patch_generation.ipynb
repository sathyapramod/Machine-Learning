```python
import matplotlib.pyplot as plt
```

```python
def create_image(rgb_img,h1,w1,h2,w2,ans):
    crop_img111 = "patch" + "041_" + ans
    crop_img1 = crop_img111 + ".png"
    crop_img11 = rgb_img[h1:h2, w1:w2]
    plt.figure()
    plt.imshow(crop_img11)
    #print(crop_img11.shape)
    cv2.imwrite(crop_img1, crop_img11)
```

```python
import cv2
```

```python
img = cv2.imread("./diaretdb1_v_1_1/resources/images/ddb1_fundusimages/image041.png")
```

```python
#cv2.imshow('image',img)
b,g,r = cv2.split(img)       # get b,g,r
rgb_img = cv2.merge([r,g,b])
r_img = cv2.merge([r])
#g_img = cv2.merge([g])
# switch it to rgb
```

```python
plt.figure()
plt.imshow(rgb_img)
plt.figure()
plt.imshow(r_img)
```

```python
h,w = r_img.shape
#print(h)
#print(w)
```

```python
count = 0
for i in range(0,w,250):
    for j in range(0,h,250):
        h1 = j
        h2 = j+250
        #print(h1)
        #print(h2)
        w1 = i
        w2 = i+250
        #print(w1)
        #print(w2)
        #print(h1)
        #print(h2)
        create_image(r_img,h1,w1,h2,w2,str(count))
        count+=1
#print(count)
```
